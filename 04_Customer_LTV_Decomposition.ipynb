Customer LTV Decomposition: Probabilistic Lifetime Value Modelling

**Business Context:** Understanding customer lifetime value (LTV) is critical for strategic decisions in acquisition, retention, and resource allocation. This model decomposes LTV into probabilistic components—purchase frequency, average order value, and retention probability—to simulate customer value trajectories.

**Question:** How can we estimate customer LTV using probabilistic modelling to account for uncertainty in behaviour?

**Approach:**
- Model purchase frequency using a Poisson distribution
- Model average order value using a Gamma distribution
- Model retention probability using a Beta distribution
- Simulate LTV across customer segments
- Visualize distribution and define strategic thresholds

 **Keywords:** Customer LTV, probabilistic modelling, retention, purchase frequency, order value

```python
# --- Notebook Setup & Imports ---

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- Parameters ---
np.random.seed(42)
N_CUSTOMERS = 1000

# Purchase frequency: Poisson (λ = 5 purchases/year)
purchase_freq = np.random.poisson(lam=5, size=N_CUSTOMERS)

# Average order value: Gamma (shape=2, scale=50)
order_value = np.random.gamma(shape=2, scale=50, size=N_CUSTOMERS)

# Retention probability: Beta (α=2, β=5)
retention_prob = np.random.beta(a=2, b=5, size=N_CUSTOMERS)
```

```python
# --- LTV Calculation ---

# Assume LTV = purchase_freq * order_value * retention_prob
ltv = purchase_freq * order_value * retention_prob

# --- Summary Statistics ---
mean_ltv = np.mean(ltv)
median_ltv = np.median(ltv)
percentile_90 = np.percentile(ltv, 90)

# --- Plot LTV Distribution ---
plt.figure(figsize=(10, 6))
sns.histplot(ltv, bins=50, kde=True, color='skyblue')
plt.axvline(mean_ltv, color='blue', linestyle='--', label=f'Mean LTV: {mean_ltv:.2f}')
plt.axvline(median_ltv, color='green', linestyle='--', label=f'Median LTV: {median_ltv:.2f}')
plt.axvline(percentile_90, color='red', linestyle='--', label=f'90th Percentile: {percentile_90:.2f}')
plt.title('Customer Lifetime Value Distribution')
plt.xlabel('LTV')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

--- Executive Summary & Strategic Recommendation ---

**Strategic Insight:**

- **LTV** varies significantly across customers due to behavioral uncertainty  
- **Probabilistic** decomposition reveals hidden value in high-retention, high-frequency segments
- **90th Percentile LTV** is substantially higher than the mean, indicating the potential for targetted scaling 
- **Recommendation** Use probabilistic LTV modelling for segmentation and targetting, prioritise retention strategies for high-frequency, high value customers, and integrate LTV simulations into acquisition cost planning and ROI forecasting
